```python
import math

import matplotlib.pyplot as plt

import numpy as np

%matplotlib inline

fout = open('scritical.txt','w')

def f1(w,k):
    return 2.0*math.atan(math.sqrt((w-k)/(w+k))) - math.sqrt(w*w - k*k)*(1.0/4.0) 

def f2(w,k):
    return 2.0*math.atan(math.sqrt((w+k)/(w-k))) - math.sqrt(w*w - k*k)*(1.0/4.0)

def f3(w,k):
    return math.log(abs(math.sqrt(k+w)+math.sqrt(k-w))/abs(math.sqrt(k+w)-math.sqrt(k-w))) - math.sqrt(k*k - w*w)*(1.0/4.0)

eps = 0.000010

def df1(w,k):
    return 1.0/(1.0+(w-k)/(w+k))*(-1.0+(k/4.0))*(math.pow(w+k,-1.0/2.0)*math.pow(w-k,-1.0/2.0)+math.pow(w+k,-3.0/2.0)*math.pow(w-k,1.0/2.0))

def df2(w,k):
    return 1.0/(1.0+(w+k)/(w-k))*(1.0+(k/4.0))*(math.pow(w+k,-1.0/2.0)*math.pow(w-k,-1.0/2.0)+math.pow(w+k,1.0/2.0)*math.pow(w-k,-3.0/2.0))

def df3(w,k):
    return (1.0-k/4.0)*math.pow(k*k - w*w,-1.0/2.0)


for i in range (100,2000,1):
    
    w = i * 0.0010 * 2 * math.pi
    
    k1 = abs(w - 2*math.pi)
    
   # if k1 > w:
        
        #plt.scatter(w,k1,c="red")
    
    #else:
        
        #plt.scatter(w,k1,c="blue")
    
    if w > 2 * math.pi and w < 4 * math.pi:
            
        k = 3.5
        
        while True:
            
            k -= f2(w,k)/df2(w,k) 
            
            if abs(f2(w,k)) < eps:
                
                
                
                fout.write(str(w)+","+str(k/w)+'\n')
            
                break

        for j in range(1,1200,1):
            
            if j**0.01 < k:
                
                plt.scatter(w,j*0.01, c="lightgreen",s=1)
                
            else:
                
                
                plt.scatter(w,j*0.01, c="lightblue",s=1)
                
        plt.scatter(w,k, c="black",s=10)
        
                
        
            
                
             
    if w > 4 * math.pi:
            
        k = 10.0
        
        while True:
            
            k -= f2(w,k)/df2(w,k) 
            
            if abs(f2(w,k)) < eps:
                
                plt.scatter(w,k, c="black",s=10)
                
                fout.write(str(w)+","+str(k/w)+'\n')
                    
                break
                
            for j in range(1,1200,1):
            
                if j**0.01 < k:
                
                    plt.scatter(w,j*0.01, c="lightgreen",s=1)
                
                else:
                
                
                    plt.scatter(w,j*0.01, c="lightblue",s=1)
                    
            plt.scatter(w,k, c="black",s=10)
                
        
                
    if w < 2 * math.pi and w > 4:
            
        k = 1.5
        
        while True:
            
            k -= f1(w,k)/df1(w,k) 
            
            if abs(f1(w,k)) < eps:
                
                plt.scatter(w,k, c = "black",s=10)
                
                fout.write(str(w)+","+str(k/w)+'\n')
                
                break
                
        for j in range(1,1200,1):
            
            if j**0.01 < k:
                
                plt.scatter(w,j*0.01, c="lightgreen",s=1)
                
            else:
                
                
                plt.scatter(w,j*0.01, c="lightblue",s=1)
                
        plt.scatter(w,k, c="black",s=10)
                
       
                
    if w < 4:
            
        k = 5.0
        
        while True:
            
            k -= f3(w,k)/df3(w,k) 
            
            if abs(f3(w,k)) < eps:
                
                plt.scatter(w,k, c= "black",s=10)
                
                fout.write(str(w)+","+str(k/w)+'\n')
                
                break
                
        for j in range(1,1200,1):
            
            if j**0.01 < k:
                
                plt.scatter(w,j*0.01, c="lightcoral",s=1)
                
            else:
                
                
                plt.scatter(w,j*0.01, c="lightblue",s=1)
                
        plt.scatter(w,k, c="black",s=10)
                
                
fout.close()                
plt.show()              # グラフ表示
                
#plt.axes().set_aspect('equal', 'datalim')
                
```
